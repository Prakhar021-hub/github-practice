```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris 
```

```python
data=load_iris()
print(data)
data.DESCR
```

```python
df=pd.DataFrame(data.data, columns=data.feature_names)
```

```python
X=df
X
```

```python
y=data.target
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train , y_test =train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier=DecisionTreeClassifier(criterion='entropy')
```

```python
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
from sklearn import tree  ## this is the visualisation of the decision tree
plt.figure(figsize=(12,10))
tree.plot_tree(classifier, filled=True)
```

```python
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
## we have studied the concept of prepruning and postpruning to avoid overfitting
```

```python
#postpruning >> on observing the splitting we have noticed that the last split is overfitted 
```

```python
classifier=DecisionTreeClassifier(criterion='entropy', max_depth= 3)
classifier.fit(X_train , y_train)
```

```python
from sklearn import tree  ## this is the visualisation of the decision tree
plt.figure(figsize=(12,10))
tree.plot_tree(classifier, filled=True)
```

```python
y_pred=classifier.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
## prepruning with hyperparameter tuning
```

```python
parameter = {
    'criterion':['gini', 'entropy', 'log_loss'],
    'splitter':['best', 'random'],
    'max_depth' : [1,2,3,4,5,6],
    'max_features' : ['sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf= DecisionTreeClassifier()
model=GridSearchCV(clf, param_grid= parameter, cv=5 , scoring='accuracy' , verbose=3)
model
```

```python
model.fit(X_train, y_train)
```

```python
model.best_estimator_
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
y_predict=model.best_estimator_.predict(X_test)
```

```python
accuracy_score(y_predict, y_test)
```
